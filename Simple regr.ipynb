```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
#Training
#splitting data to xtrain xtest ytrain ytest train is for training and test for prediction
from sklearn.model_selection import train_test_split
#train_test_split(x,y,size,not necessary)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
#to predict answer
y_pred = regressor.predict(X_test)
```

```python
#plotting training set points
plt.scatter(X_train, y_train, color = 'red')
#plot regression line
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary Vs Exp (Training Set)')
plt.xlabel('Years of Exp')
plt.ylabel('Salary')
plt.show()
```

```python
#plotting test set points real observsion
plt.scatter(X_test, y_test, color = 'red')
#plot regression line train only as line will be same(predicted obs)
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary Vs Exp (Test Set)')
plt.xlabel('Years of Exp')
plt.ylabel('Salary')
plt.show()
```

```python
#To predict further single obs [[]] means 2D array
regressor.predict([[12]])
```

```python
#to get the equation of regression
print(regressor.coef_)
print(regressor.intercept_)
```

```python
print('Salary = ', regressor.coef_, 'Years of exp + ', regressor.intercept_)
```
